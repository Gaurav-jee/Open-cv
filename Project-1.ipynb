```python
mp_drawing = mp.solutions.drawing_utils
holistic = mp.solutions.holistic
mp_pose = mp.solutions.pose
```

```python
cap = cv2.VideoCapture(0)
with holistic.Holistic(min_detection_confidence=0.2, min_tracking_confidence=0.5) as holis:
    while 1:
        _,frame = cap.read()
        fr = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        frame = cv2.flip(frame,1)
        result = holis.process(frame)
        height, width,_ = frame.shape
        y_mid = height//2
        
        if result.right_hand_landmarks:
            right_hand = (result.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_WRIST].x * width,
                         result.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_WRIST].y * height)
            
            x=right_hand[0]
            pose = "MOVE"
            if(right_hand[1]<y_mid):
                pose="acc"
                pdi.keyUp('left')
                pdi.keyDown('right')
            elif(right_hand[1]>y_mid):
                pose="break"
                pdi.keyUp('right')
                pdi.keyDown('left')
            else:
                pose="Move"
                pdi.keyUp('right')
                pdi.keyUp('left')
        cv2.line(frame, (0,y), (width,y), (0,0,255), 2)
        cv2.putText(frame, pose, (30, 30), cv2.FONT_HERSHEY_SIMPLEX, 2, (0,255,0), 2)
        cv2.imshow('win2', frame)
        if cv2.waitKey(1)==ord('q'):
            break;
cap.release()
cv2.destroyAllWindows()
```
